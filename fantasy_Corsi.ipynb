```python
import pandas as pd
from datetime import datetime, timedelta, timezone

date_format = "%d-%m-%Y"
today = datetime.utcnow()-timedelta(days=1.5)
```

```python
file_name = f"data/summary_statsS-{today.strftime('%Y-%m-%d')}.csv"
player_biosS = pd.read_csv(file_name, index_col=False)
file_name = f"data/summary_statsG-{today.strftime('%Y-%m-%d')}.csv"
player_biosG = pd.read_csv(file_name, index_col=False)
for index, row in player_biosG.iterrows():
    player_biosG.at[index, 'tertiaryPosition'] = 'G'
player_bios = pd.concat([player_biosS, player_biosG], axis=0)
player_bios
```

```python
file_name = f"data/corsi-{today.strftime('%Y-%m-%d')}.csv"
Corsi = pd.read_csv(file_name, index_col=False)
Corsi
```

```python
player_list = player_bios['playerId'].unique()
```

```python
valid_types = ['SHOT', 'BLOCK', 'MISS', 'GOAL']
Corsi_Only = Corsi[Corsi['event'].isin(valid_types)]
Corsi_Only
```

```python
corsi_player_totals = pd.DataFrame()

for player in player_list:
#     print(player)
    temp = Corsi_Only.copy()
    
    CorsiFor = 0
    CorsiAgainst = 0
    
    team_temp = player_bios.loc[player_bios['playerId'] == player]['team'].iloc[0]
    
    temp['awayCorsi'] = Corsi_Only['awayNumbers'].apply(lambda x: str(player) in x)
    temp['homeCorsi'] = Corsi_Only['homeNumbers'].apply(lambda x: str(player) in x)
    
    temp_all_Corsi = temp[(temp['awayCorsi'] == True) | (temp['homeCorsi'] == True)]
    
    for index, row in temp_all_Corsi.iterrows():
        if row['Corsi'] == team_temp:
            CorsiFor += 1
        else:
            CorsiAgainst +=1
        
    temp_df = pd.DataFrame([{'playerId': player, 'CorsiFor': CorsiFor, 'CorsiAgainst': CorsiAgainst}])
    
    corsi_player_totals = pd.concat([corsi_player_totals, temp_df], axis=0, ignore_index=True)
    
    
```

```python
player_bios = pd.merge(player_bios, corsi_player_totals, how='left', on='playerId')
```

```python
valid_types = ['SHOT', 'BLOCK', 'HIT', 'GOAL']
fantasyCorsi = Corsi[Corsi['event'].isin(valid_types)]
fantasyCorsi
```

```python
corsi_fantasy_totals = pd.DataFrame()

for player in player_list:
#     print(player)
    temp = fantasyCorsi.copy()
    
    CorsiFor = 0
    CorsiAgainst = 0
    
    team_temp = player_bios.loc[player_bios['playerId'] == player]['team'].iloc[0]
    
    temp['awayCorsi'] = Corsi_Only['awayNumbers'].apply(lambda x: str(player) in x)
    temp['homeCorsi'] = Corsi_Only['homeNumbers'].apply(lambda x: str(player) in x)
    
    temp_all_Corsi = temp[(temp['awayCorsi'] == True) | (temp['homeCorsi'] == True)]
    
    print(player_bios.loc[player_bios['playerId'] == player]['name'].iloc[0])
#     print(len(temp_all_Corsi))
#     print('--------')
    
    for index, row in temp_all_Corsi.iterrows():
        if row['event'] == 'HIT':
            if row['Corsi'] == team_temp:
                CorsiFor += 0.1
            else:
                CorsiAgainst += 0.1
        elif row['event'] == 'GOAL':
            if row['strength'] == 'EV':
                if row['Corsi'] == team_temp:
                    CorsiFor += 3.7
                else:
                    CorsiAgainst += 3.7
            else:
                if row['Corsi'] == team_temp:
                    CorsiFor += 4.2
                else:
                    CorsiAgainst += 4.2
        elif row['event'] == 'BLOCK':
            if row['Corsi'] == team_temp:
                CorsiAgainst += 0.5
            else:
                CorsiFor += 0.5
        elif row['event'] == 'SHOT':
            if row['Corsi'] == team_temp:
                CorsiFor += 0.1
            else:
                CorsiAgainst += 0.1

        
    temp_df = pd.DataFrame([{'playerId': player, 'FantasyCorsiFor': CorsiFor, 'FantasyCorsiAgainst': CorsiAgainst}])
    
    corsi_fantasy_totals = pd.concat([corsi_fantasy_totals, temp_df], axis=0, ignore_index=True)
```

```python
fantasy_bios = pd.merge(player_bios, corsi_fantasy_totals, how='left', on='playerId')
fantasy_bios
```

```python
file_name = f"data/all_df_fp-{today.strftime('%Y-%m-%d')}.csv"
all_df = pd.read_csv(file_name, index_col=False)
all_df

def summary_statistics(df):
    temp_build = df.groupby('playerId').agg({
        'name': 'first',
        'team': 'last',
        'position': 'first',
        'toi': 'sum',
        'gamesPlayed': 'sum',
        'goals': 'sum',
        'assists': 'sum',
        'specialTeams': 'sum',
        'shots': 'sum',
        'hits': 'sum',
        'blockedShots': 'sum',
        'powerPlayToi': 'sum',
        'shorthandedToi': 'sum',
        'plusMinus': 'sum',
        'fantasyPoints': 'sum',
        'secondaryPosition': 'first',
        'tertiaryPosition': 'first'
    }).reset_index()



    temp_build['FPP60'] = (temp_build['fantasyPoints'] / temp_build['toi'] * 3600).round(2)
    temp_build['FPPG'] = (temp_build['fantasyPoints'] / temp_build['gamesPlayed']).round(2)

    temp_build = temp_build.sort_values(by='FPPG', ascending=False)
    
    return temp_build
```

```python
vatrano = all_df.loc[all_df['name'] == 'F. Vatrano']
vatrano = vatrano.loc[vatrano['gameDate'] > '2023-10-28']
summary_statistics(vatrano)
```
